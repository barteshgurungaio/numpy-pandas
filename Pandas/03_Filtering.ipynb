### Filtering 
* Approach 1: Logical Operators ` : == > < <= >= != & |  isin([]) `
* Approach 2: df.`query('expr')`

```python
import pandas as pd
```

```python
df = pd.read_csv('data/customer_filtering.csv')
df.head(10)
```

```python
df.shape
```

```python
# Approach 1: logical operators
df[df['country'] == 'UK']

df[df['country'] != 'UK']

df[df['transaction_amount'] < 500]

df[df['transaction_amount'] <= 300]

df[df['transaction_amount'] >= 800]

df[(df['country']=='UK') & (df['transaction_amount'] <= 300)]  

df[(df['country']=='UK') & (df['transaction_amount'] <= 300)] [['contact_name', 'transaction_amount']] # filter & see certain columns

df[(df['country']=='UK') | (df['transaction_amount'] == 800)].iloc[0:5] # to see only 5 records: also done with head()

# Note: [] means no records
```

```python
df[df['city'].isin(['Buenos Aires', 'Graz', 'Bruxelles'])]
```

```python
# Approach 2: df.query
df.query('transaction_amount >= 800 & country == "Brazil" | (country == "UK" and transaction_amount <= 300)')
```

### Filtering with String functions
`str.contains   str.startswith   str.endswith   ~ (Tilde)`

```python
df[df['contact_name'].str.startswith('R')]

df[df['contact_name'].str.endswith('y')]

df[df['contact_name'].str.contains('Hardy')]

df[~(df['country'] == 'Argentina')] # ~ means Except Argentina
```

```python
df['contact_name'].str.upper()
df['contact_name'].str.lower()
df['contact_name'].str.capitalize()
df['contact_name'].str.split(' ')
df['contact_name'].str.len()
df['contact_name'].str.strip() # lstrip  rstrip
df['contact_name'].str.title()
df['contact_name'].str.find('a')
df['contact_name'].str.replace(' ', ',')
df['contact_name'].str.count('a')
df['contact_name'].str.isnumeric()
df['contact_name'].str.isalnum()
df['contact_name'].str.isalpha()
```
