* 1. drop the rows where `all columns` are duplicate
* 2. drop the rows where `one column` is duplicate
* 3. drop the rows where `multiple column` are duplicate
* 4. drop the rows where `one/multiple column` are duplicate and keep first/last

```python
import pandas as pd
df = pd.read_csv('data/remove_dup.csv')
df
```

```python
# 1. `all columns` are duplicate
df.drop_duplicates()
```

```python
# 2. 'one column' is duplicate 
df.drop_duplicates('City')
```

```python
# 3. 'multiple columns' are duplicate 
df.drop_duplicates(['Country', 'City'])
```

```python
# 4. `one/multiple column` are duplicate and keep first/last
df.drop_duplicates(['Country', 'City'], keep='first')
```
