### Pivot Table: Excellent way to Summarize your Data!

```python
import pandas as pd
```

```python
df = pd.read_excel('data/loan_sample.xlsx')
```

```python
df.head()
```

```python
df.pivot_table(
    index='Profession', 
    values='Income',
    aggfunc='count'           
               ).head()

# Note: aggfunction: default is mean/avg
# If you don't specify any col, it will take up the numeric column
```

```python
df.pivot_table(
    index='Profession', 
    columns='Gender',
    values='Income',
    aggfunc='count'           
               ).head()
```

```python
df.pivot_table(
    index='Profession', 
    columns='Gender',
    values='Income',
    aggfunc='count',
    fill_value=0           
               ).head()
```

```python
df.pivot_table(
    index='Profession', 
    columns=['Gender','Own_House'],
    values='Income',
    aggfunc=['count','sum'],
    fill_value=0           
               ).head()
```

```python
df.pivot_table(
    index='Profession', 
    columns=['Gender','Own_House'],
    values='Income',
    aggfunc=['count','sum'],
    fill_value=0,
    margins=True           
               )
```
